```python
import pandas as pd
import numpy as np
import statistics as stat
from matplotlib import pyplot as plt
import random
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.neighbors import KNeighborsClassifier
from imblearn.over_sampling import SMOTE
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization
from tensorflow.keras import optimizers
```

```python
ucl = pd.read_csv("ucl_stats.csv")
ucl = ucl.drop('team', axis = 1)
x = ucl.drop('champions', axis=1)
y = ucl['champions']
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = .25)
```

```python
smt = SMOTE()
x_train, y_train = smt.fit_resample(x_train, y_train)
```

```python
iswinner = y_train == 1
champions = x_train[iswinner]
isloser = y_train == 0
losers = x_train[isloser]
```

```python
# Average goals scored for the two dfs
champions_average_goals_scored = stat.mean(champions['goals_scored'])
losers_average_goals_scored = stat.mean(losers['goals_scored'])

# Average goals conceded for the two dfs
champions_average_goals_conceded = stat.mean(champions['goals_conceded'])
losers_average_goals_conceded = stat.mean(losers['goals_conceded'])

# Average goal difference for the two dfs
champions_average_goal_difference = stat.mean(champions['gd'])
losers_average_goal_difference = stat.mean(losers['gd'])

# Average wins for the two dfs
champions_average_wins = stat.mean(champions['wins'])
losers_average_wins = stat.mean(losers['wins'])

# Average draws for the two dfs
champions_average_draws = stat.mean(champions['draws'])
losers_average_draws = stat.mean(losers['draws'])

# Average losses for the two dfs
champions_average_losts = stat.mean(champions['losts'])
losers_average_losts = stat.mean(losers['losts'])

# Average matches playes for the two dfs
champions_average_matches_played = stat.mean(champions['match_played'])
losers_average_matches_played = stat.mean(losers['match_played'])

# Standard Deviation for goals scored
champions_sd_goals_scored = stat.stdev(champions['goals_scored'])
losers_sd_goals_scored = stat.stdev(losers['goals_scored'])

# Standard Deviation for the goals conceded
champions_sd_goals_conceded = stat.stdev(champions['goals_conceded'])
losers_sd_goals_conceded = stat.stdev(losers['goals_conceded'])

# Standard Deviaiton for the goal difference
champions_sd_goal_difference = stat.stdev(champions['gd'])
losers_sd_goal_difference = stat.stdev(losers['gd'])

# Standard Deviation for the wins
champions_sd_wins = stat.stdev(champions['wins'])
losers_sd_wins = stat.stdev(losers['wins'])

# Standard Deviation for the draws
champions_sd_draws = stat.stdev(champions['draws'])
losers_sd_draws = stat.stdev(losers['draws'])

# Standard Deviation for the losses
champions_sd_losts = stat.stdev(champions['losts'])
losers_sd_losts = stat.stdev(losers['losts'])

# Standard Deviation for the matches played
champions_sd_matches_played = stat.stdev(champions['losts'])
losers_sd_matches_played = stat.stdev(losers['match_played'])
```

```python
# df to display the summary statistics

# Goals scored stats
goals_scored_stats = pd.DataFrame(data = [[champions_average_goals_scored, champions_sd_goals_scored], [losers_average_goals_scored, losers_sd_goals_scored]], columns = ["Mean", "Standard Deviation"])
goals_scored_stats = goals_scored_stats.rename(index = {0:"Champions", 1:"Losers"})
print("Goals Scored Stats")
print(goals_scored_stats)
print("\n")

# Goals Conceded Stats
goals_conceded_stats = pd.DataFrame(data = [[champions_average_goals_conceded, champions_sd_goals_conceded], [losers_average_goals_conceded, losers_sd_goals_conceded]], columns = ["Mean", "Standard Deviation"])
goals_conceded_stats = goals_conceded_stats.rename(index = {0:"Champions", 1:"Losers"})
print("Goals Conceded Stats")
print(goals_conceded_stats)
print("\n")

# Goal difference Stats
goal_difference_stats = pd.DataFrame(data = [[champions_average_goal_difference, champions_sd_goal_difference], [losers_average_goal_difference, losers_sd_goal_difference]], columns = ["Mean", "Standard Deviation"])
goal_difference_stats = goal_difference_stats.rename(index = {0:"Champions", 1:"Losers"})
print("Goal Difference Stats")
print(goal_difference_stats)
print("\n")

# Win stats
win_stats = pd.DataFrame(data = [[champions_average_wins, champions_sd_wins], [losers_average_wins, losers_sd_wins]], columns = ["Mean", "Standard Deviation"])
win_stats = win_stats.rename(index = {0:"Champions", 1:"Losers"})
print("Win Stats")
print(win_stats)
print("\n")

# Draw stats
draw_stats = pd.DataFrame(data = [[champions_average_draws, champions_sd_draws], [losers_average_draws, losers_sd_draws]], columns = ["Mean", "Standard Deviation"])
draw_stats = draw_stats.rename(index = {0:"Champions", 1:"Losers"})
print("Draw Stats")
print(draw_stats)
print("\n")

# Losts stats
losts_stats = pd.DataFrame(data = [[champions_average_losts, champions_sd_losts], [losers_average_losts, losers_sd_losts]], columns = ["Mean", "Standard Deviation"])
losts_stats = losts_stats.rename(index = {0:"Champions", 1:"Losers"})
print("Losts Stats")
print(losts_stats)
print("\n")

# Matches Played stats
matches_stats = pd.DataFrame(data = [[champions_average_matches_played, champions_sd_matches_played], [losers_average_matches_played, losers_sd_matches_played]], columns = ["Mean", "Standard Deviation"])
matches_stats = matches_stats.rename(index = {0:"Champions", 1:"Losers"})
print("Matches Played Stats")
print(matches_stats)
print("\n")

```

```python
# KNN model

features = ['match_played', 'wins', 'draws','losts', 'goals_scored', 'goals_conceded', 'gd', 'group_point']
data = x_train
testData = x_test
target = y_train

# kNN where k = 1
model = KNeighborsClassifier(1)

# preprocessing may be needed on target to convert strings to ints
model.fit(data, target)

# make predictions
expected = target
predicted = model.predict(data)


# summarize the fit of the model
# support: the number of occurrences of each class
# macro average: averaging the unweighted mean per label
# weighted average: averaging the support-weighted mean per label
print("kNN where k = 1:\n")
print("Train data:")
print(metrics.classification_report(expected, predicted, target_names= ['champions', 'losers']))
print("Confusion Matrix:")
print(metrics.confusion_matrix(expected, predicted))
print("\n")

print("Test data:")
expected = y_test
predicted = model.predict(testData)
print(metrics.classification_report(expected, predicted, target_names= ['champions', 'losers']))
print("Confusion Matrix:")
print(metrics.confusion_matrix(expected, predicted))
```
